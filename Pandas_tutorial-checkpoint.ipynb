## filna() . This fuction is used to fill the na box 

### parse_dates = ['column_name'] is used to convert other datatype to dates

```python
dff = pd.read_csv("weather_data1.csv",parse_dates=['day'])
dff
```

```python
dff.set_index('day',inplace=True)
```

```python
dff
```

#### Now in this dataframe, we have several null values or NaN values. If we want to fill those values with any specific values.. for eg:0,use fillna() function 

```python
new_df = dff.fillna(0)
new_df
```

#### But filling in all the columns is not an appropriate task. For example in the above df,  event column cant have 0 as event. so we can pass specific dictionaries to fillna()  

```python
new_df = dff.fillna({
    'temperature':0,
    'windspeed':0,
    'event': 'No Event'
})
new_df
```

##### But filling 0 in the missing boxes is not a good practice. Because if there is any chance to take mean, the values will differ in a large extend . And it is quite impossible that first day the temperature in 30 and the next day is drop to 0 and the next day is 28. so there must be a trend

### 1. one method is that we can copy the previous column value to the next. For that use fillna() with an attribute method=ffill. ffill means forward fill 

```python
new_df = dff.fillna(method='ffill')
new_df
```

### 2. if we need to reverse this process, that is if we want to fill the box with the values next to it, we can use backfill. method=bfill 

```python
new_df = dff.fillna(method='bfill')
new_df
```

#### 3. if we add anothere attribute axis='columns' , this fillna() process will happen horizontally. By default axis is row, so the row value will copy the value in the adjascent row, if we give axis = columns, then it will do it column wise. 

```python
new_df = dff.fillna(method='ffill', axis='columns')
new_df
# forward fill
```

```python
new_df = dff.fillna(method='bfill',axis='columns')
new_df
# backfill
```

#### in the date 8 and 9 two rows of temperature are NaN. In dates 7,8,9 three rows of windspeed is NaN. If we need to do ffill or bfill and we want to limit that only one row should be done, not more than one row is allowed to do bfill or ffill, we can set limit through limit attribute

```python
dff.set_index('day',inplace=True)
dff
```

```python
new_df = dff.fillna(method='ffill',limit=1)
new_df
```

```python
new_df = dff.fillna(method='ffill',limit=2)
new_df
```

# interpolate() function
### this function will fill the nan boxes with much more realistic values. The values will be more close related to the existing nearest values. Some methods in interpolate() are:
#### 1. Linear- by default, method = Linear.  filling missing values with a linearly interpolated value between the two nearest data points.
#### 2.'nearest': Fills missing values with the nearest non-null value. works same as fillna() bfill or ffill
#### 3. quadratic
#### 4.cubic
#### 5.time

```python
new_df = dff.interpolate()
new_df
```

```python
new_df = dff.interpolate(method='nearest')
new_df
```

```python
new_df = dff.interpolate(method='quadratic')
new_df
```

```python
new_df = dff.interpolate(method='cubic')
new_df
```

```python
new_df = dff.interpolate(method='time')
new_df
```

# dropna() is used to drop the column of row which has na value

```python
new_df = dff.dropna()
new_df
```

#### parameters:
#### 1. axis-- if 0 or index , it is for row and if 1 or columns it if for column. default is 0

```python
new_df = dff.dropna(axis=1)
new_df
```

#### 2.how-- how = any: deletes if any empty box, how=all: delete if all the boxes are empty, by default how=any. In this example no row or column got deleted because there is no cell which all the elements are empty

```python
new_df = dff.dropna(how='all')
new_df
```

### threshold - we can give a threshold value. if there is a row or column has atleast threshold value of non-na value . it will not get droped. suppose if the threshold value is 1 so a row having atleast 1 value will not get dropped and row which has no value will get deleted.
